```python
df=pd.read_csv("Spam.csv")
df.head()
```

```python
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis='columns',inplace=True)
df.head()
```

```python
df.groupby('v1').describe()
```

```python
df['Spam']=df['v1'].apply(lambda x:1 if x== 'spam' else 0)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.v2,df.Spam,test_size=0.25)
```

```python
v=CountVectorizer()
X_train_count=v.fit_transform(X_train.values)
```

```python
print(X_train_count[1])
```

```python
type(X_train_count.toarray())
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count.toarray(),y_train)
```

```python
email = ['Congratulations. You have won 10000 USD']
email_count=v.transform(email)
model.predict(email_count)
```
